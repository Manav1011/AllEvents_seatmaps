```python
data_str = open('current.json','r').read()
```

```python
import json
import re
data = json.loads(data_str)
seat_arr = {}
for i in data['seatmap']:
    if(i['status'] == 'blocked'):
        alpha = re.findall('[a-zA-Z]', i['name'])        
        if alpha[0] in seat_arr:
            seat_arr[alpha[0]].append(i['name'])
        else:
            seat_arr[alpha[0]] = [i['name']]
print(json.dumps(seat_arr,indent=3)) 
```
